```python
from lxml import etree
from requests import get
import sys
import os
```

```python
URL = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&field=title&term=\"{}\""
xpath = "//Id"
```

```python
import csv
title_list = []
with open('/content/drive/MyDrive/Research stuff/bib_to_csv.csv') as csvfile:
  data = csv.reader(csvfile)
  for row in data:
    title_list.append(row[0].strip())
print(title_list)
```

```python
def xml_id_extractor(page):
   #extract the ids from page and return that as string
   content = etree.fromstring(page)
   root = content.getroottree()
   result = [item.text for item in root.xpath(xpath)]
   if len(result)<1:
       return "0"
   return result
```

```python
i=0
with open('/content/drive/MyDrive/Research stuff/ids_bib.csv','w') as f1:
  writer=csv.writer(f1, lineterminator='\n')
  for title in title_list:
    try:
      page = get(URL.format(title)).content
      id = xml_id_extractor(page)
      writer.writerow(id)
      i= i + 1
      print(i)
    except:
      continue
```
