```python
from bs4 import BeautifulSoup
import csv
import urllib.request
```

```python
pmids = []
with open('/content/drive/MyDrive/Research stuff/ids_bib.csv') as csvfile:
  data = csv.reader(csvfile)
  for row in data:
    if (row[0].strip() != '0'):
      pmids.append(row[0].strip())
print(len(pmids))
```

```python
with open('/content/drive/MyDrive/Research stuff/output_positive.csv','w') as f1, open('/content/drive/MyDrive/Research stuff/output_positive_not_found.csv','w') as f2:
 writer1=csv.writer(f1, lineterminator='\n',)
 writer2= csv.writer(f2, lineterminator='\n',)
 i = 0
 for id in pmids:
    try:
      soup = BeautifulSoup(urllib.request.urlopen('http://www.ncbi.nlm.nih.gov/pubmed/'+ id.rstrip()).read())
      abstract = soup.find_all(attrs={"class": "abstract-content"})
      abstract = (abstract[0].p.string)
      writer1.writerow([abstract])
      print (i)
    except:
        writer2.writerow([id])
    
```
